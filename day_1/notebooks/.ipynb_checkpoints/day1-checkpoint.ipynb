```python
# modules to use
import pandas as pd
import numpy as np

# read in all the data
nfl_data = pd.read_csv("../input/NFL Play by Play 2009-2017 (v4).csv")
sf_permits = pd.read_csv("../input/Building_Permits.csv")

# set seed for reproducibility
np.random.seed(0)
```

```python
nfl_data.sample(5)
```

```python
sf_permits.sample(5)
```

# See how many missing data points we have

```python
missing_values_count = nfl_data.isnull().sum()
missing_values_count[0:10]
```

```python
total_cells = np.product(nfl_data.shape)
total_missing = missing_values_count.sum()

# percentage of data that is missing
(total_missing*1.0/total_cells) * 100
```

```python
missing_vals_count_sf = sf_permits.isnull().sum()
missing_vals_count_sf.loc[["Street Number Suffix", "Zipcode"]]
```

```python
total_cells_sf = np.product(sf_permits.shape)
total_missing_sf = missing_vals_count_sf.sum()

# percentage of data missing
(total_missing_sf*1.0/total_cells_sf)*100
```

# Drop missing values

```python
nfl_data.dropna()
```

```python
columns_with_na_dropped = nfl_data.dropna(axis=1)
columns_with_na_dropped.head()
```

```python
print "Columns in original data", nfl_data.shape[1]
print "Columns in na's dropped: %d" % columns_with_na_dropped.shape[1]
```
